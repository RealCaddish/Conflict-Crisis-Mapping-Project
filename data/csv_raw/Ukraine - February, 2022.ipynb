```python
import pandas as pd
import geopandas as gpd 
import matplotlib.pyplot as plt
```

```python
# import ACLED data from file
data = gpd.read_file('../csv_raw/February_2022.csv')
```

```python
# what are the datatypes, column values, and memory usage?
data.info()
```

```python
# explore rows and columns
data.head()
```

```python
# create geodataframe object from lat, long points
gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data.longitude, data.latitude))
```

```python
gdf.plot()
```
